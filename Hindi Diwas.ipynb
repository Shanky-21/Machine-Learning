```python
#Import Library
import numpy as np
import matplotlib.pyplot as plt
import re
from PIL import Image
```

```python
# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS
```

```python
stopword=open("stopwords.txt",'r', encoding = "utf-8")
stop_words=[]
```

```python
#pre-process stopword
for i in stopword:
    i = re.sub('[\n]', '', i)
    stop_words.append(i)
```

```python
stopwords = set(stop_words)
```

```python
text = open('hindi diwas.txt', 'r',encoding = "utf-8").read()
print("File successfully read")
print(len(text))
```

```python
# save mask to alice_mask
mask = np.array(Image.open('indmap2.jpg'))
    
print('Image imported')
```

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python

# instantiate a word cloud object
Hindi_wc = WordCloud(font_path = "Lohit-Devanagari.ttf",
                     background_color='Black', max_words=2000, mask=mask, stopwords=stopword)

# generate the word cloud
Hindi_wc.generate(text)

# display the word cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(Hindi_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
